## Customer-level feature engineering

Now that we have a cleaned transaction-level dataset, it's time to **roll it up** to the customer level, which we'll feed into our machine learning algorithms later.

We want 1 customer per row, and we want the features to represent information such as:

- Number of unique purchases by the customer
- Average cart value for the customer
- Total sales for the customer
- Etc.

To do so, we'll use two tools we've seen already:

1. <code style="color:steelblue">.groupby()</code> to roll up by customer.
2. <code style="color:steelblue">.agg()</code> to engineer aggregated features.

<br>
**First, we'll aggegrate invoice data by customer. We'll engineer 1 feature:**
* <code style="color:steelblue">'total_transactions'</code> - the total number of unique transactions for each customer.

```python
# Aggegrate invoice data
invoice_data = df.groupby('CustomerID').InvoiceNo.agg({ 'total_transactions' : 'nunique' })

# Display invoice data for first 5 customers
invoice_data.head()
```

Essentially, we counted the number of unique Invoice ID's for each customer.

- First, we grouped by CustomerID.
- Then, to get number of unique transactions, we used the Invoice ID column.
- Finally, we passed in a dictionary of aggregations to the .agg() function.
- **nunique** counts the number of unique values within each group.

**Next, using the same technique, aggregate product data by customer. Engineer 2 features:**
* <code style="color:steelblue">'total_products'</code> - the total number of products purchased by each customer (not the unit Quantity within each product!).
* <code style="color:steelblue">'total_unique_products'</code> - the number of unique products purchased by each customer.
* We will want to use the Item ID column (i.e. StockCode) after our groupby!
* Let's name it <code style="color:crimson">product_data</code>.

```python
# Aggregate product data
product_data = df.groupby('CustomerID').StockCode.agg( { 'total_products' : 'count', 
                                                     'total_unique_products' : 'nunique' } )

# Display product data for first 5 customers
product_data.head()
```

**Finally, aggregate sales data by customer. Engineer 2 features:**
* <code style="color:steelblue">'total_sales'</code> - the total sales for each customer.
* <code style="color:steelblue">'avg_product_value'</code> - the average value of the products purchased by the customer (not the UnitPrice!).
* We will want to use the Sales column after your groupby!
* Let's name it <code style="color:crimson">sales_data</code>.

```python
# Roll up sales data
sales_data = df.groupby('CustomerID').Sales.agg( { 'total_sales' : 'sum', 
                                                  'avg_product_value' : 'mean' } )

# Display sales data for first 5 customers
sales_data.head()
```

## Intermediary levels

You won't always be able to easily roll up to customer-level directly... Sometimes, it will be easier to create intermediary levels first.

<br>
**Let's first aggregate cart data at the "cart-level."**
* We'll group by <code style="color:steelblue">'CustomerID'</code> AND by <code style="color:steelblue">'InvoiceID'</code>. Remember, we're treating each invoice as a "cart."
* Then, we'll calculate <code style="color:steelblue">'cart_value'</code> by taking the sum of the Sales column. This is the total sales by invoice (i.e. cart).
* Finally, we'll call <code style="color:steelblue">.reset_index()</code> to turn CustomerID and InvoiceID back into regular columns so we can perform another aggregation.

```python
# Aggregate cart-level data (i.e. invoice-level)
cart_data = df.groupby(['CustomerID' , 'InvoiceNo']).Sales.agg( { 'cart_value' : 'sum' })

# Reset index
cart_data.reset_index(inplace=True)

# Display cart data for first 5 CARTS
cart_data.head()
```

**Aggregate cart data by customer. Engineer 3 features:**
* <code style="color:steelblue">'avg_cart_value'</code> - average cart value by customer.
* <code style="color:steelblue">'min_cart_value'</code> - minimum cart value by customer.
* <code style="color:steelblue">'max_cart_value'</code> - maximum cart value by customer.
* Let's name it <code style="color:crimson">agg_cart_data</code>.

```python
# Aggregate cart data (at customer-level)
agg_cart_data = cart_data.groupby('CustomerID').cart_value.agg( { 'avg_cart_value' : 'mean', 
                                                                 'min_cart_value' : 'min',
                                                                 'max_cart_value' : 'max'})

# Display cart data for first 5 CUSTOMERS
agg_cart_data.head()
```

## Joining together the ABT

```python
# Join together customer-level ABT
customer_df = invoice_data.join([product_data, sales_data, agg_cart_data])

# Display customer-level data for first 5 customers
customer_df.head()
```

## Thresholds

One very **simple and straightforward way** to reduce the dimensionality of this item data is to set a **threshold** for keeping features.

The rationale is that you might only want to keep popular items.
For example, let's say item A was only purchased by 2 customers. Well, the feature for item A will be 0 for almost all observations, which isn't very helpful.
On the other hand, let's say item B was purchased by 100 customers. The feature for item B will allow more meaningful comparisons.

To make this concrete, assume we only wish to keep item features for the 20 most popular items.

<br>
**First, we can see which items those are and the number of times they were purchased.**
1. Take the sum by column.
* Sort the values.
* Look at the last 20 (since they are sorted in ascending order by default).

```python
# Display most popular 20 items
item_data.sum().sort_values().tail(20)
```

**Next, if we take the <code style="color:steelblue">.index</code> of the above series, we can get just a list of the StockCodes for those 20 items.**

**Finally, we can keep only the features for those 20 items.**

```python
# Keep only features for top 20 items
top_20_item_data = item_data[top_20_items]

# Shape of remaining dataframe
top_20_item_data.shape
```

```python
# Display first 5 rows of top_20_item_data
top_20_item_data.head()
```

## Toy example: oval blob

PCA is an elegant technique that's very practical, but it can be a bit hard to try it on our larger dataset right away.

Instead, let's first use another toy example to break down the intuition behind PCA.

<br>
**First, we'll create an "oval blob" synthetic dataset.**
* Instead of both input features and a target variable, we'll only have 2 input features.
* $x_1$ will be random samples from a normal distribution.
* $x_2$ will be random samples from a normal distribution **plus $x_1$**.

```python
# Set random seed
np.random.seed(101)

# Create first feature: x1
x1 = np.random.normal(0, 1, 100)

# Create second feature: x2
x2 = x1 + np.random.normal(0, 1, 100)
```

**Let's stack them together to form <code style="color:steelblue">X</code>, our input feature matrix.**
* We'll skip putting them in a Pandas DataFrame.
* Instead, we'll just stack them together as two separate columns in a matrix.

```python
# Stack together as columns
X = np.stack([x1, x2], axis=1)

# Print shape of X
print( X.shape )
```

Now, one of the key assumptions for PCA is that the features should be on the same scale.

**Create a scaled version of our feature matrix and name it <code style="color:steelblue">X_scaled</code>.**
* We'll use use <code style="color:steelblue">StandardScaler</code>.

```python
# Initialize instance of StandardScaler
scaler = StandardScaler()

# Fit and transform X
X_scaled = scaler.fit_transform(X)

# Display first 5 rows of X_scaled
X_scaled[:5]
```

**Plot a scatterplot of the scaled $x_1$ against the scaled $x_2$.**
* We will label the axes as <code style="color:crimson">x1 (scaled)</code> and <code style="color:crimson">x2 (scaled)</code>.
* We will call <code style="color:steelblue">plt.axis('equal')</code> after the scatterplot to put plot's axes on the same scale.

```python
# Plot scatterplot of scaled x1 against scaled x2
plt.scatter(X_scaled[:,0], X_scaled[:,1])

# Put plot axes on the same scale
plt.axis('equal')

# Label axes
plt.xlabel('x1 (scaled)')
plt.ylabel('x2 (scaled)')

# Clear text residue
plt.show()
```

## Principal components
***Principal Component Analysis (PCA)*** is an Unsupervised Learning task that creates a sequence of new, uncorrelated features that each try to maximize its "explained variance" of the original dataset.

- It does so by generating linear combinations from the original features.
- These new features are meant to replace the original ones.

We initialize an instance of the PCA transformation.**
* Then, we can fit the instance with just one line of code.

```python
from sklearn.decomposition import PCA

# Initialize instance of PCA transformation
pca = PCA()

# Fit the instance
pca.fit(X_scaled)
```

Once fitted, the pca instance now contains the principal components.

- These are the coefficients of the linear combinations of our original features.
- Furthermore, they define "directions" that are uncorrelated with each other.
- Each component in the sequence is along the "direction" that captures the most variance, after accounting for the previous components.


**Let's display the principal components.**

```python
# Display principal components
pca.components_
```

- The top row, [ 0.70710678,  0.70710678], is the first principal component
- The second row, [-0.70710678,  0.70710678], is the second principal component.

Since we only have 2 original features, we can have at most 2 principal components. The maximum possible number of principal components is either the number of original features or the number of observations in your dataset, whichever is lower.

**Next, let's overlay the principal components on our oval blob.**
* We'll make the oval blob semi-transparent so we can see the principal components more easily.
* For clarity, we'll plot the first principal component in **black**, and we'll make it twice as long.
* We'll plot the 2nd principal component in <strong style="color:red">red</strong>.

```python
# Plot scaled dataset and make it partially transparent
plt.scatter(X_scaled[:,0], X_scaled[:,1], alpha=0.3)

# Plot first principal component in black
plt.plot([0, 2*pca.components_[0,0]], [0, 2*pca.components_[0,1]], 'k')

# Plot second principal component in red
plt.plot([0, pca.components_[1,0]], [0, pca.components_[1,1]], 'r')

# Set axes
plt.axis('equal')
plt.xlabel('x1 (scaled)')
plt.ylabel('x2 (scaled)')

# Clear text residue
plt.show()
```

As you can see, the first principal component **(in black)** represents the "direction" of the most variance in the dataset.

- This is the along the **"length"** of the oval.
- Because the length is on a diagonal, neither x1 nor x2 can capture this direction alone.
- The principal component, as a linear combination of x1 and x2, can capture that direction.

The second principal component **(in red)** represents the direction of the most variance in the dataset after accounting for the first principal component.

- This is the along the **"width"** of the oval.
- Again, the width is on a diagonal, so neither x1 nor x2 can capture that direction alone.

**Finally, we can transform our original features using the fitted PCA instance.**
* Remember, PCA generates new features that **replace** the original ones.
* We'll name these new features <code style="color:crimson">PC</code> (short for "principal component") to indicate that they are generated from PCA.

```python
# Generate new features
PC = pca.transform(X_scaled)
```

Now, if we were to plot a scatterplot of these "principal component features," we can see that we've captured all of the variance of the original dataset.

```python
# Plot transformed dataset
plt.scatter(PC[:,0], PC[:,1], alpha=0.3, color='g')

# Plot first principal component in black
plt.plot([0, 2], [0, 0], 'k')

# Plot second principal component in red
plt.plot([0, 0], [0, 1], 'r')

# Set axes
plt.axis('equal')
plt.xlabel('PC1')
plt.ylabel('PC2')

# Clear text residue
plt.show()
```

A few important things to point out:

- Essentially, we've **"rotated"** that original oval blob.
- Now, the length and width of the oval are no longer on diagonals.
- PC1 alone is able to capture the direction of greatest variance.
- And when we keep both PC1 and PC2, we can completely recreate the original dataset.


## Dimensionality reduction

Here's where dimensionality reduction comes into play. You can just keep some number of the "best" components, a.k.a. the ones that explain the most variance.

- Remember, PCA creates a sequence of principal components and each one tries to capture the most variance after accounting for the ones before it.
- So we know that our first principal component captured the most variance, but how much did it actually capture?

<br>
**First, display the "explained variance" ratio of each component.**

```python
# Display explained variance ratio
pca.explained_variance_ratio_
```

So the first principal component captured almost 87% of the total variance in the dataset!

- The second component only captured about 13% of the total variance.
- That means we can just keep the first component and still capture most of the variance.

**To keep only 1 component, we can simply initialize and fit another PCA transformation.**
* This time, we'll simply pass the argument <code style="color:steelblue">n_components=1</code>.

```python
# Initialize and fit a PCA transformation, only keeping 1 component
pca = PCA(n_components=1)
pca.fit(X_scaled)
```

Now, we only have the first component.

As you can see, we only kept the length of the oval.

We've essentially projected our 2-dimensional data into 1 dimension.

And the best part? We still captured 87% of the original variance!

## K-Means
nb: Typically, the best clustering model is the one that creates clusters that are intuitive and reasonable in the eyes of the key stakeholder (i.e. our client, the online gift retailer). Therefore, which algorithm we choose is typically less important than the input features that you feed into it.

### Intuition of K-Means

Pseudocode:

1. You declare a hyperparameter `k`, the number of different clusters you want.
2. The algorithm initializes k random points in your feature space (called centroids).
3. For each observation in your dataset, its distance to each centroid is calculated.
4. Each observation is assigned to the cluster of its closest centroid.
5. Then, each centroid is shifted to the mean of the observations assigned to that cluster.
6. Steps (3) - (5) are repeated until the centroids no longer shift.

Bottom line: this process produces k clusters that "clump together" observations based on distance.

### Feature sets
Because K-Means creates clusters based on distances, and because distances are calculated by between observations defined by their feature values, **the features you choose to input into the algorithm heavily influence the clusters that are created.**

For this spcedific project, let's look at 3 possible feature sets and compare the clusters created from them. We'll try:

1. Only purchase pattern features ("Base DF")

Again, it's worth reminding that there aren't any clear, widely-accepted performance metrics for clustering tasks because we don't have labeled "ground truth."

- For regression, you could calculate the error between your predictions and the true values.
- For classification, you could calculate the AUROC of your predictions and the true classes.
- For clustering, you don't know what's "true."

Therefore, it's usually more useful to just compare the clusters created by different approaches.

**Let's import the adjusted Rand index from Scikit-Learn.**

```python
# Adjusted Rand index
from sklearn.metrics import adjusted_rand_score
```

The **adjusted Rand index** is a similarity measure between two clusterings. It essentially counts the observations that were assigned to the same or different clusters by two different clusterings.

- Values close to 0.0 indicate random independent clusters
- Values close to 1.0 indicate nearly identical clusters

**For example, we can see the adjusted Rand index between <code style="color:steelblue">base_df.cluster</code> and <code style="color:steelblue">threshold_df.cluster</code> like so:**

```python
# Similary between base_df.cluster and threshold_df.cluster
adjusted_rand_score(base_df.cluster, threshold_df.cluster)
```

Furthermore, the adjusted Rand index is **symmetric**, which means you can pass in the clusters in any order.

```python
# Similary between threshold_df.cluster and base_df.cluster
adjusted_rand_score(threshold_df.cluster, base_df.cluster)
```

**Finally, display the adjusted Rand index between <code style="color:steelblue">base_df.cluster</code> and <code style="color:steelblue">pca_df.cluster</code>.**
* Is it higher or lower than the one between <code style="color:steelblue">base_df.cluster</code> and <code style="color:steelblue">threshold_df.cluster</code>?
* Which clusters are more like the ones created from the the base dataframe?

```python
# Similary between base_df.cluster and pca_df.cluster
adjusted_rand_score(base_df.cluster, pca_df.cluster)
```
